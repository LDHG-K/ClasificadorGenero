los descriptores son los siguientes:

* ancho de la frente cm - Esta columna está en CM's. Este es el ancho de la frente.

* ancho de la nariz - Esta columna contiene 0's y 1's donde 1 es "nariz ancha" y 0 es "nariz no ancha".

* nariz larga - Esta columna contiene 0 y 1 donde 1 es "nariz larga" y 0 es "nariz no larga".

* distancenosetoliplong - Esta columna contiene 0's y 1's donde 1 representa la "larga distancia entre la nariz y los labios" mientras que 0 es "corta distancia entre la nariz y los labios".


```
import numpy
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('gender_classification_v7.csv')  


etiqueta = df.drop(['long_hair','forehead_width_cm','forehead_height_cm','nose_wide','nose_long','lips_thin','distance_nose_to_lip_long'],axis=1)
atributos = df.drop(['gender'],axis=1)
df.head()
atributos.head()
```

```
pos_hombre=df.gender=="Male"
pos_mujer=df.gender=="Female"



alturaFrente1 = df.forehead_height_cm[pos_hombre]
alturaFrente0 = df.forehead_height_cm[pos_mujer]

plt.figure()
plt.plot(alturaFrente1,'*',label="Hombres")
plt.plot(alturaFrente0,'.',label="Mujeres")
plt.grid()
plt.legend()
plt.ylabel("Alto de la frente")
plt.show()

```

## en esta grafica se identifica que las mujeres tienden a tener un alto de frente maximo de 6.50 y los hombres mas de 7.00

```
anchoFrente1 = df.forehead_width_cm[pos_hombre]
anchoFrente0 = df.forehead_width_cm[pos_mujer]

plt.figure()
plt.plot(anchoFrente1,'*',label="Hombres")
plt.plot(anchoFrente0,'.',label="Mujeres")
plt.grid()
plt.legend()
plt.ylabel("Ancho de la frente")
plt.show()
```

##en esta grafica se identifica que las mujeres tienden a tener un ancho de frente maximo de 14.4 y los hombres mas de 15

## Clasificacion

```
clasificador = tree.DecisionTreeClassifier()
clasificador = clasificador.fit(atributos,etiqueta)

print(clasificador.predict([[1,14.1,6.2,1,0,1,1]]))
```
